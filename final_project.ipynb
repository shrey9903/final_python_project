```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("jacksoncrow/stock-market-dataset")

print("Path to dataset files:", path)
```

```python
path
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os

# List the contents of the directory to find the CSV file
files = os.listdir(path)
print("Files in the dataset directory:", files)

# Assuming there's one main CSV file, or picking the first one ending with .csv
csv_file = None
for f in files:
    if f.endswith('.csv'):
        csv_file = os.path.join(path, f)
        break

if csv_file:
    df = pd.read_csv(csv_file)
    print("Dataset loaded successfully!")
    print(df.head())
else:
    print("No CSV file found in the directory.")
```

```python
print("Exchange Wise Companies")

exchange_avg=df.groupby("Listing Exchange")["Round Lot Size"].mean()
exchange_avg
```

```python
print("Market Category Wise Companies")

market_avg=df.groupby("Market Category")["Round Lot Size"].mean()
market_avg
```

```python
print("ETF Wise Distribution")

etf_avg=df.groupby("ETF")["Round Lot Size"].mean()
etf_avg
```

```python
exchange_counts=df["Listing Exchange"].value_counts()

plt.figure(figsize=(8,5))
plt.bar(exchange_counts.index, exchange_counts.values, color='skyblue', edgecolor='black')

plt.title("Companies by Listing Exchange")
plt.xlabel("Exchange")
plt.ylabel("Number of Companies")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df["Round Lot Size"], df.index, color='red')

plt.title("Round Lot Size Distribution")
plt.xlabel("Round Lot Size")
plt.ylabel("Index")

plt.show()
```

```python
sorted_df = df.sort_values("Round Lot Size")

plt.figure(figsize=(8,5))

plt.plot(sorted_df["Round Lot Size"])

plt.title("Trend of Round Lot Size")
plt.xlabel("Index")
plt.ylabel("Round Lot Size")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Round Lot Size"], bins=15, edgecolor='black')

plt.title("Distribution of Round Lot Size")
plt.xlabel("Round Lot Size")
plt.ylabel("Frequency")

plt.show()
```

```python
etf_counts = df["ETF"].value_counts()

plt.figure(figsize=(6,6))

plt.pie(etf_counts.values, labels=etf_counts.index, autopct="%1.1f%%")

plt.title("ETF vs Non ETF Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

df.boxplot(column="Round Lot Size", by="Listing Exchange")

plt.title("Round Lot Size by Exchange")
plt.suptitle("")
plt.xlabel("Exchange")
plt.ylabel("Round Lot Size")

plt.show()
```

```python
correlation = df.corr(numeric_only=True)

plt.figure(figsize=(8,6))

sns.heatmap(correlation,
            annot=True,
            cmap="coolwarm",
            fmt=".2f",
            linewidths=0.5)
plt.title("Correlation Matrix of Numerical Features")
plt.xlabel("Features")
plt.ylabel("Features")

plt.tight_layout()

plt.show()
```
